```python
# data_frames
# NOTE:Health Screen Size extremely low
df_blood = pd.read_csv("./blood_donations.csv")
df_blood.name = "df_blood"
df_blood_state = pd.read_csv("./blood_donations_state.csv")
df_blood_state.name = "df_blood_state"
df_health_screen = pd.read_csv("./infant_immunisation.csv")
df_health_screen.name = "df_health_screen"
df_organ_pledges = pd.read_csv("./pekab40_screenings_state.csv")
df_organ_pledges.name = "df_organ_pledges"
df_covid_deaths = pd.read_csv("./covid_cases.csv")
df_covid_deaths.name = "df_covid_deaths"
df_covid_cases = pd.read_csv("./covid_deaths_linelist.csv")
df_covid_cases.name = "df_covid_cases"
df_list = [
    df_blood,
    df_blood_state,
    df_health_screen,
    df_organ_pledges,
    df_covid_cases,
    df_covid_deaths,
]
```

```python
# Data Cleaning and Duplicate/Null Value Identification
for _ in df_list:
    null_count = _.isnull().sum().sum()
    duplicate_count = _.isnull().sum().sum()
    if null_count > 0:
        print(f"{_.name} has null values")
        print(null_count)
    elif duplicate_count > 0:
        print(f"{_.name} has duplicate values")
        print(duplicate_count)
    else:
        print(f"{_.name} is clean")
```
