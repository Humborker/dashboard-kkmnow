```python
def boxplot(df, x_col, y_col, title="Distribution Analysis"):
    fig = px.box(
        df,
        x=x_col,
        y=y_col,
        points="outliers",
        color_discrete_sequence=["#636EFA"],  # Fixed brand color
    )

    fig.update_layout(
        title={"text": title, "x": 0.05, "xanchor": "left"},
        plot_bgcolor="rgba(0,0,0,0)",
        paper_bgcolor="white",
        font=dict(family="Arial, sans-serif", size=14, color="#2c3e50"),
        margin=dict(l=40, r=40, t=60, b=40),
        xaxis=dict(showgrid=False, linecolor="#d6d6d6"),
        yaxis=dict(showgrid=True, gridcolor="#eeeeee"),
    )

    return fig


def lineplot(df, x_col, y_col, title="Trend Analysis"):
    fig = px.line(
        df, x=x_col, y=y_col, markers=True, color_discrete_sequence=["#EF553B"]
    )

    fig.update_layout(
        title={"text": title, "x": 0.05, "xanchor": "left"},
        hovermode="x unified",  # Best UX for comparing multiple lines
        plot_bgcolor="rgba(0,0,0,0)",
        paper_bgcolor="white",
        font=dict(family="Arial, sans-serif", size=14, color="#2c3e50"),
        margin=dict(l=40, r=40, t=60, b=40),
        xaxis=dict(showgrid=False, linecolor="#d6d6d6", ticks="outside"),
        yaxis=dict(showgrid=True, gridcolor="#eeeeee"),
    )

    # Improve line smoothing and thickness
    fig.update_traces(line=dict(width=3), marker=dict(size=8))

    return fig


def barplot(df, x_col, y_col, title="Comparative Analysis"):
    """
    Standardized Bar Plot.
    UI/UX: Minimalist vertical bars, no x-grid, clean data separation.
    """
    fig = px.bar(df, x=x_col, y=y_col, color_discrete_sequence=["#00CC96"])

    fig.update_layout(
        title={"text": title, "x": 0.05, "xanchor": "left"},
        plot_bgcolor="rgba(0,0,0,0)",
        paper_bgcolor="white",
        font=dict(family="Arial, sans-serif", size=14, color="#2c3e50"),
        margin=dict(l=40, r=40, t=60, b=40),
        xaxis=dict(showgrid=False, linecolor="#d6d6d6"),
        yaxis=dict(showgrid=True, gridcolor="#eeeeee"),
        bargap=0.2,  # Fixed UX spacing between bars
    )

    fig.update_traces(marker_line_width=0)  # Keeps it modern and flat

    return fig


def scatterplot(df, x_col, y_col, title="Correlation Study"):
    """
    Standardized Scatter Plot.
    UI/UX: Semi-transparent markers to reveal data density/clusters.
    """
    fig = px.scatter(df, x=x_col, y=y_col, color_discrete_sequence=["#AB63FA"])

    fig.update_layout(
        title={"text": title, "x": 0.05, "xanchor": "left"},
        hovermode="closest",
        plot_bgcolor="rgba(0,0,0,0)",
        paper_bgcolor="white",
        font=dict(family="Arial, sans-serif", size=14, color="#2c3e50"),
        margin=dict(l=40, r=40, t=60, b=40),
        xaxis=dict(showgrid=True, gridcolor="#f0f0f0", linecolor="#d6d6d6"),
        yaxis=dict(showgrid=True, gridcolor="#f0f0f0", linecolor="#d6d6d6"),
    )

    fig.update_traces(
        marker=dict(size=10, opacity=0.7, line=dict(width=1, color="White"))
    )

    return fig


def histogram(df, x_col, title="Frequency Distribution"):
    """
    Standardized Histogram.
    UI/UX: Unified binning, clean y-axis for frequency count.
    """
    fig = px.histogram(df, x=x_col, color_discrete_sequence=["#FFA15A"])

    fig.update_layout(
        title={"text": title, "x": 0.05, "xanchor": "left"},
        bargap=0.05,  # Slight gap between bins for visual clarity
        plot_bgcolor="rgba(0,0,0,0)",
        paper_bgcolor="white",
        font=dict(family="Arial, sans-serif", size=14, color="#2c3e50"),
        margin=dict(l=40, r=40, t=60, b=40),
        xaxis=dict(showgrid=False, linecolor="#d6d6d6"),
        yaxis=dict(showgrid=True, gridcolor="#eeeeee", title="Count"),
    )

    return fig
```

```python
# CLI Functions


def column_selector(df, axis_name):
    columns = df.columns.tolist()

    print(f"\n--- Select column for {axis_name} ---")
    for i, col in enumerate(columns):
        print(f"{i}. {col}")

    while True:
        try:
            choice = int(input(f"Enter the number for {axis_name}: "))
            if 0 <= choice < len(columns):
                return columns[choice]
            print("Invalid number. Try again.")
        except ValueError:
            print("Please enter a valid integer.")
```

```python
# Primary CLI
def main():

    print("--- KKMNOW Dashboard Generator ---")

    # Dictionary for holding parquet urls
    url_dict = {
        "organ pledges": "https://storage.data.gov.my/healthcare/organ_pledges_state.parquet",
        "covid cases": "https://storage.data.gov.my/healthcare/covid_cases_age.parquet",
        "blood donations": "https://storage.data.gov.my/healthcare/blood_donations_state.parquet",
        "covid deaths": "https://storage.data.gov.my/healthcare/blood_donations_state.parquet",
    }
    # Dictionary for holding plotly functions
    plot_map = {
        "1": {"label": "Bar Chart", "func": barplot},
        "2": {"label": "Scatter Plot", "func": scatterplot},
        "3": {"label": "Line Chart", "func": lineplot},
        "4": {"label": "Box Plot", "func": boxplot},
    }
    # CLI prompts for choosing and downloading parquet urls with case correction
    while True:
        url = input("""
        Choose one of the following options to analyse:
        A) organ pledges 
        B) covid cases
        C) blood donations
        D) covid deaths
        """).strip()
        # Error handling and loading of parquet into dataframe variable
        if url not in url_dict:
            print("Please provide a correct answer.")
        else:
            try:
                # Loading Parquet directly into Pandas
                df = pd.read_parquet(url_dict[url])
                print(f"Success! Loaded {len(df)} rows.")
                x_col = column_selector(df, "X-Axis")
                y_col = column_selector(df, "Y-Axis")
                break
            except Exception as e:
                print(f"Failed to load data{e}")
    # Plot prompt for creating plotly from url
    while True:
        plot = input("""
        Choose one of the following plots to create:
        1) barplot 
        2) scatterplot
        3) lineplot 
        4) boxplot 
        """).strip()
        if plot in plot_map:
            graph_fun = plot_map[plot]["func"]
            fig = graph_fun(df, x_col, y_col)
            with open("template.html") as f:
                j2_template = Template(f.read())
            with open("plotly_graph.html", "w", encoding="utf-8") as f:
                f.write(j2_template.render(fig=fig.to_html(full_html=False)))
            file_path = "plotly_graph.html"
            break

    # Ensure the file exists and is accessible
    if os.path.isfile(file_path):
        # Open the file in the default browser
        webbrowser.open(f"file://{os.path.realpath(file_path)}")
    else:
        print(f"File not found: {file_path}")


if __name__ == "__main__":
    main()
```

```python
# Time Series/Feature Engineering Prediction
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer


def create_feature_pipeline(df, target_column):
    """
    Creates a scikit-learn pipeline for feature engineering.
    Automatically separates numerical and categorical features.
    """
    # 1. Identify feature types (excluding the target)
    X = df.drop(columns=[target_column])

    numeric_features = X.select_dtypes(include=["int64", "float64"]).columns
    categorical_features = X.select_dtypes(include=["object", "category"]).columns

    # 2. Define preprocessing 
245/245 ━━━━━━━━━━━━━━━━━━━━ 2s 4ms/step - loss: 285.6737 - mae: 7.0061 - val_loss: 85.8856 - val_mae: 5.0038
Epoch 7/20
245/245 ━━━━━━━━━━━━━━━━━━━━ 1s 3ms/step - loss: 27.5337 - mae: 2.2329 - val_loss: 15.6148 - val_mae: 2.2050
Epoch 8/20
245/245 ━━━━━━━━━━━━━━━━━━━━ 1s 3ms/step - loss: 331.8246 - mae: 5for numerical data
    # We impute missing values with the mean and then scale
    numeric_transformer = Pipeline(
        steps=[
            ("imputer", SimpleImputer(strategy="mean")),
            ("scaler", StandardScaler()),
        ]
    )

    # 3. Define preprocessing for categorical data
    # We impute missing values with a constant and then one-hot encode
    categorical_transformer = Pipeline(
        steps=[
            ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
            ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
        ]
    )

    # 4. Combine transformers into a ColumnTransformer
    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, numeric_features),
            ("cat", categorical_transformer, categorical_features),
        ]
    )

    # 5. Create the full pipeline (you can add a classifier/regressor here later)
    full_pipeline = Pipeline(steps=[("preprocessor", preprocessor)])

    return full_pipeline
```

```python
pipeline = create_feature_pipeline(df, "target")
X_processed = pipeline.fit_transform(df.drop(columns=["target"]))
y = df["target"].values

# 2. Initialize the model with the correct input size
# X_processed.shape[1] gives us the number of features after One-Hot Encoding
input_size = X_processed.shape[1]
model = create_keras_model(input_dim=input_size, task="classification")

# 3. Fit the model
model.fit(X_processed, y, epochs=10, batch_size=32)
```
